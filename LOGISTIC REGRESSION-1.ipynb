```python
df.groupby('Pclass').mean()['Age']#mean of age according to pclass
```

```python
def func1(list):
    Age=list[0]
    Pclass=list[1]
    if pd.isnull(Age):
        if Pclass==1:
            return 38
        elif Pclass==2:
            return 29
        else:
            return 25
    else:
        return Age
```

```python
df['Age']=df[['Age','Pclass']].apply(func1,axis=1)
```

```python
s=pd.get_dummies(df['Sex'],drop_first=True)
```

```python
t=pd.get_dummies(df['Embarked'],drop_first=True)
```

```python
df.drop(['male','Q'],axis=1,inplace=True)#column not correlated
```

```python
X=df.drop('Survived',axis=1)
```

```python
y=df['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
a=LogisticRegression()
```

```python
a.fit(X_train,y_train)
```

```python
y_pred=a.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_pred,y_test)
```
